### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X=house_s.drop(columns=['price'])
y=house_s.price
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=18)
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

### Build a base model

```python
import statsmodels.api as sm

# Evaluation Metrics....

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
X_train_c= sm.add_constant(X_train)
X_test_c= sm.add_constant(X_test)
```

```python
base_model= sm.OLS(y_train, X_train_c).fit()
print(base_model.summary())
```

```python
print('Statistically significant columns: ')
for i in base_model.pvalues.to_dict():
    if base_model.pvalues.to_dict()[i]<0.05:
        print(i)
```

```python
base_model.params[1:].sort_values().plot(kind='bar')
```

```python
def validation(fitted_model,train_x,train_y,test_x,test_y):
    y_pred_test = fitted_model.predict(test_x)
    r2 = r2_score(test_y,y_pred_test)
    print('R2',r2)
    adj_r2 = 1- (((1-r2)*(train_x.shape[0]-1))/(train_x.shape[0] - train_x.shape[1] - 1))
    print('Adjusted R2',adj_r2)
    print('MSE',mean_squared_error(test_y,y_pred_test))
    print('RMSE',mean_squared_error(test_y,y_pred_test)**0.5)
    print('MAPE',mean_absolute_percentage_error(test_y,y_pred_test))    
```

```python
validation(base_model, X_train_c, y_train, X_test_c, y_test)
```

### Model from Sklearn

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
def model_validation(model,train_x,train_y,test_x,test_y):
    global m
    m = model
    m.fit(X_train,y_train)
    print('%s'%model)
    print('Train Parameters:')
    validation(m,train_x,train_y,train_x,train_y)
    print('\nTest parametrs:')
    validation(m,train_x,train_y,test_x,test_y)
```

```python
model_validation(LinearRegression(),X_train,y_train,X_test,y_test)
```

### Decision Tree

```python
model_validation(DecisionTreeRegressor(),X_train,y_train,X_test,y_test)
```

```python
# as we can see the model is overfitted.....
```

### Random Forest

```python
model_validation(RandomForestRegressor(),X_train,y_train,X_test,y_test)
```

### AdaBoost Regressor

```python
model_validation(AdaBoostRegressor(),X_train,y_train,X_test,y_test)
```

### Gradient Boost Regressor

```python
model_validation(GradientBoostingRegressor(),X_train,y_train,X_test,y_test)
```

### XGBoost Regressor

```python
model_validation(XGBRegressor(),X_train,y_train,X_test,y_test)
```

**As we can see that the gradient boost gives the best R2 value and there is no overfitting in the gradient boost regressor, so we can go with the gradient boost regressor as the final model for prediction.**
